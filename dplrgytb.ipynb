```python
import numpy as np
import matplotlib.pyplot as plt

X= np.array([-12,-10,-4,-7,1,3,6,9,15,24,32,38])
y = np.array([3,6,9,12,15,18,20,25,29,40,55,58])
plt.scatter(X,y)
```

```python
#input output shapes
house_info = tf.constant(["bedroom",'bathroom','garage'])
house_price = tf.constant([939700])
house_info, house_price
```

```python
#turn numpy array into tensors
x= tf.constant(X)
y1=tf.constant(y)
x,y1
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    tf.keras.Input(shape=(1,)),
    Dense(50,activation='relu'),
    Dense(1)
])

model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(learning_rate=0.002),
                      metrics = ['mae'])
model.fit(x,y1,epochs=100)
```

```python
x_test=tf.constant([12,15,20,25,70])
y_test=model.predict(x_test)
```

```python
plt.scatter(x_test,y_test)
plt.scatter(x,y,color='r',marker='+')
```
